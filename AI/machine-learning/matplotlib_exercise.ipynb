```python
# Function for nth fibonacci number 
def fibonacci(n):
	a = 0
	b = 1
	
	# Check is n is less
	# than 0
	if n < 0:
		print("Incorrect input")
		
	# Check is n is equal
	# to 0
	elif n == 0:
		return 0
	
	# Check if n is equal to 1
	elif n == 1:
		return b
	else:
		for i in range(1, n):
			c = a + b
			a = b
			b = c
		return b

temp_arr = np.arange(0,1000)


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(temp_arr,list(map(fibonacci, temp_arr)), label='fib')
ax.legend(loc='best')
# ax.set_xlim([0,100])
```

```python
temp_arr_fib = np.array()
for i in temp_arr:
    temp_arr_fib.append(fibonacci(i))
```
